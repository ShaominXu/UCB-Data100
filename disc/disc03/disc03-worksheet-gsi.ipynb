## Problem 1

Which of the following lines of code returns a `Series` with the mean vote percentage for each political party, for all years given, sorted in **decreasing** order? Choose all that apply.

    A. elections.groupby('Party')['%'].agg('mean').sort_values()
    
    B. elections.groupby('Party')['%'].agg('mean').sort_values(ascending = False)
    
    C. elections.groupby('Party')['%'].mean().sort_values()
    
    D. elections.groupby('Party')['%'].mean().sort_values(ascending = False)

*Answer*: B, D

```python
# A: no!, it's in increasing order
elections.groupby('Party')['%'].agg('mean').sort_values()
```

```python
# B: yes!
elections.groupby('Party')['%'].agg('mean').sort_values(ascending = False)
```

```python
# C: no!, in increasing order once again
elections.groupby('Party')['%'].mean().sort_values()
```

```python
# D: yes!
elections.groupby('Party')['%'].mean().sort_values(ascending = False)
```

Write a line of `pandas` code that returns a `Series` with the year as the index, and the total number of votes that were cast across all parties for that year. What would you change in your code so the output is a `DataFrame` object?

```python
# Solution 1
elections.groupby('Year')['Popular vote'].agg(sum)
```

## Problem 3

Write a line of `pandas` code that returns a `Series` whose index are the years and whose values are the number of candidates that participated in those years' elections.

*Answer:*

```python
elections.groupby('Year').size()
```

<span style="color:blue">
Staff Notes: We often get questions about using <code>.count()</code> instead of <code>.size()</code>. <code>.size()</code> includes rows with <code>NaN</code> values in the aggregation and returns a <code>Series</code> rather than a <code>DataFrame</code>, as the <code>size()</code> is the same for all columns. On the other hand, <code>count()</code> is called per column, as results can differ depending on the number of <code>NaNs</code>/missing values, and a <code>DataFrame</code> is returned</span>

## Problem 4

Write a line of `pandas` code that creates a filtered `DataFrame` named `filtered_parties`
from the elections dataset and keeps only the parties that have at least one election %
more than 50%.

*Answer*: 

```python
filtered_parties = elections.groupby("Party").filter(lambda sf: sf["%"].max() >= 50)
filtered_parties
```

<span style="color:blue">
Staff Notes: Remember to iterate the differences between filter and agg. Filter does
NOT change the final granularity of the DataFrame that is returned. The index will
not be the groups like in groupby + agg. Walking through the way that filter is similar
in application to groupby (the applying of a function) and contrasting it with the ways
that are different (filter functions needing to return a boolean) can be really helpful to
disentangle the nuance between all these different functions.</span>

## Problem 5

Write a line of `pandas` code that uses the `filtered_parties` `DataFrame` to return a new
`DataFrame` with row indices that correspond to the year and columns that correspond
to each party. Each entry should be the total percentage of votes for all the candidates
that ran during that particular year for the specified party. Missing values from the dataset (the cases where a party did not have a candidate in a particular year) should be entered as 0. Below is an example.

![](pivot.png)

```python
elections_pivot = filtered_parties.pivot_table(
    index="Year",
    columns="Party",
    values=["%"],
    aggfunc=np.sum,
    fill_value = 0
)
elections_pivot
```

<span style="color:blue">
Staff Notes: It is not obvious at first that this is meant to be a pivot table call. Make sure
to emphasize the utility that pivot has to give us an alternative view of data. By default,
the years and party combos are not the original granularity of the data. <code>groupbys</code> AND
<code>pivots</code> help by changing granularities. Also, if desired you can tie it back to <code>groupby</code>s
by seeing if students can create the same outputs using a 2-key groupby.
